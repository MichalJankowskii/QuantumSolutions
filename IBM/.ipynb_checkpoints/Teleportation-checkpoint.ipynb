```python
import qiskit
from qiskit import *
```

```python
circuit = QuantumCircuit(3,3)
%matplotlib inline
circuit.draw(output='mpl')
```

```python
#Define quantum stat that should be teleported from q0 to q2
circuit.x(0)
#circuit.x(0) #2x not -> 1
#circuit.h(0) #hadamar state
circuit.barrier()
circuit.draw(output='mpl')
```

```python
'Teleportation protocol'
circuit.h(1)
circuit.cx(1,2)
circuit.draw(output='mpl')
```

```python
circuit.cx(0,1)
circuit.h(0)
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.measure([0,1], [0,1])
circuit.draw(output='mpl')
```

```python
circuit.barrier()
circuit.cx(1,2)
circuit.cz(0,2)
circuit.draw(output='mpl')
```

```python
'q0 teleported to q2'
circuit.measure(2, 2)
simulator = Aer.get_backend('qasm_simulator')
result = execute(circuit, backend = simulator, shots = 1024).result()
counts = result.get_counts()
from qiskit.tools.visualization import plot_histogram
plot_histogram(counts)
```
